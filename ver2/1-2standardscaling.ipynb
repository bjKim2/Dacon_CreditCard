```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../Data/train.csv")
val_df = pd.read_csv("../Data/val.csv")
test_df = pd.read_csv("../Data/test.csv")
```

```python
# 머신러닝 모듈 사용하기
# sklearn의 preprocessing기능을 사용하면 최대값과 최소값을 비교하면서 정규화가 진행

from sklearn import preprocessing


#최대값과 최소값의 값 가져오기
standard = preprocessing.StandardScaler()
standard.fit(train.iloc[:,1:])
train_df_norm = standard.transform(train.iloc[:,1:])
val_df_norm = standard.transform(val_df.iloc[:,1:31])
test_df_norm = standard.transform(test_df.iloc[:,1:31])



train_x = pd.DataFrame(train_df_norm,columns=train.columns[1:])
val_x = pd.DataFrame(val_df_norm,columns=train.columns[1:])
test_x = pd.DataFrame(test_df_norm,columns=train.columns[1:])

```

```python
val_x['Class'] = val_df['Class']
```
