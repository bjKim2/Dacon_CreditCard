```python

import pandas as pd
import numpy as np
train_norm = pd.read_csv("../norm_Data/train_norm.csv")
val_norm = pd.read_csv('../norm_Data/val_norm.csv')
test_norm = pd.read_csv('../norm_Data/test_norm.csv')
# test_df.drop('ID',axis=1,inplace=True)

train_x = train_norm.iloc[:,:-2]
val_x = val_norm.iloc[:,:-3]
val_y = val_norm['Class']
test_x = test_norm.iloc[:,:-2]
```

```python
val_x
```

```python
valid_normal = val_norm[val_norm.Class == 0]
valid_fraud = val_norm[val_norm.Class == 1]
## functions
def ztest(feature):
    mean = valid_normal[feature].mean()
    std = valid_fraud[feature].std()
    zScore = (valid_fraud[feature].mean() - mean) / (std/np.sqrt(sample_size))

    return zScore

sample_size = len(valid_fraud)
significant_features = []
critical_value = 2.58

cols = list(valid_normal.columns)
for i in cols:
  z_vavlue = ztest(i)

  if (abs(z_vavlue) >= critical_value):
    significant_features.append(i)

significant_features
```

```python
## PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=15,whiten=False)

pca.fit(train_x)
X_train_PCA = pca.transform(train_x)
X_train_PCA = pd.DataFrame(data=X_train_PCA, index=train_x.index)

X_train_PCA_inverse = pca.inverse_transform(X_train_PCA)
X_train_PCA_inverse = pd.DataFrame(data=X_train_PCA_inverse, index=train_x.index,columns= train_x.columns)
X_train_PCA_inverse


X_valid_pca = pca.transform(val_x)
X_valid_PCA = pd.DataFrame(data=X_valid_pca, index=val_x.index)

X_valid_PCA_inverse = pca.inverse_transform(X_valid_PCA)
X_valid_PCA_inverse = pd.DataFrame(data=X_valid_PCA_inverse, index=val_x.index, columns= train_x.columns)

X_test_pca = pca.transform(test_x)
X_test_PCA = pd.DataFrame(data=X_test_pca, index=test_x.index )

X_test_PCA_inverse = pca.inverse_transform(X_test_PCA)
X_test_PCA_inverse = pd.DataFrame(data=X_test_PCA_inverse, index=test_x.index, columns= train_x.columns)

X_valid_PCA['Class'] = val_y
X_valid_PCA_inverse['Class'] = val_y
```

```python
X_train_PCA.to_csv("../pca_Data/train_pca.csv",index=False)
X_train_PCA_inverse.to_csv("../pca_Data/train_inpca.csv",index=False)
X_valid_PCA.to_csv("../pca_Data/val_pca.csv",index=False)
X_valid_PCA_inverse.to_csv("../pca_Data/val_inpca.csv",index=False)
X_test_PCA.to_csv("../pca_Data/test_pca.csv",index=False)
X_test_PCA_inverse.to_csv("../pca_Data/test_inpca.csv",index=False)

```
