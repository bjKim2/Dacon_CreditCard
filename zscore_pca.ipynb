```python

import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("./norm_Data/train_norm.csv")
val_df = pd.read_csv('./Data/val.csv')
test_df = pd.read_csv('./Data/test.csv')
test_df.drop('ID',axis=1,inplace=True)
```

```python
train_x = train_df.drop(['V29','V30'],axis=1)
val_x = val_df.drop(['V29','V30','Class'],axis=1)
val_y = val_df.loc[:,'Class']
test_x = test_df.drop(['V29','V30'],axis=1)
```

```python
valid_normal = val_df[val_df.Class == 0]
valid_fraud = val_df[val_df.Class == 1]
## functions
def ztest(feature):
    mean = valid_normal[feature].mean()
    std = valid_fraud[feature].std()
    zScore = (valid_fraud[feature].mean() - mean) / (std/np.sqrt(sample_size))

    return zScore

sample_size = len(valid_fraud)
significant_features = []
critical_value = 2.58

cols = list(valid_normal.columns)
for i in cols:
  z_vavlue = ztest(i)

  if (abs(z_vavlue) >= critical_value):
    significant_features.append(i)
```
